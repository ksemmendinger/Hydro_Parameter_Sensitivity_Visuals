# Read Hydroinformatics Team Data into Pandas Dataframe

```python
# Import Statements
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import scipy
```

```python
# reads in NWM output file (model)
df_streamflow_cms = pd.read_csv('NWM_data_20190701.csv') 
plt.figure()
plt.plot(df_streamflow_cms['TimeFromStart']/(60*60*24),df_streamflow_cms['q_cms'],'r-',label='q_cms')
plt.title('Station ID:'+str(df_streamflow_cms['station'][0]))
plt.xlabel('days')
plt.ylabel('streamflow (cms)')
plt.tight_layout()
```
